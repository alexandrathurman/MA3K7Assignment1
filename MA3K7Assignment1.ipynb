## Artful Specialisation - specific sequences

This code determines if the sequence with the starting pair (1,1) eventually contains the elements (1,1).

```python
def seq1():
    # Set the starting values to be (1,1)
    a, b = 1, 1

    while True:
        # Compute the next values using our pattern
        a, b = b, (a + b) % 10

        # Check if we have returned to (1,1)
        if (a, b) == (1, 1):
            return "The starting values (1,1) form a bracelet"
    return "The sequence does not return to (1,1)"

print(seq1())
```

This code plots the bracelet generated by (1,1).

```python
import matplotlib.pyplot as plt
import numpy as np

def generate_sequence(a, b, max_length=1000):
    sequence = [a, b]
    for _ in range(max_length):
        sequence.append((sequence[-1] + sequence[-2]) % 10)
        if sequence[-2:] == sequence[:2]:
            break
    return sequence[:-2]

def plot_circle(sequence):
    n = len(sequence)
    x = [np.cos(2*np.pi*i/n) for i in range(n)]
    y = [np.sin(2*np.pi*i/n) for i in range(n)]
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.plot(x, y, 'o-')
    for i in range(n):
        ax.annotate(str(sequence[i]), (x[i], y[i]))
    ax.set_aspect('equal', 'box')
    plt.show()


sequences = []
sequence = generate_sequence(1,1)
pair = (1,1)
sequences.append((pair,sequence))

# Plot the first sequence
print(f"Bracelet for starting pair: {sequences[0][0]}")
plot_circle(sequences[0][1])

```

## Systematic Specialisation - tries every possible starting pair

This code works logically through every possible starting pair to see if the sequence eventually returns to the starting pair.

```python
def seq3(a, b):
    # Store the original pair
    starting_pair = (a, b)

    while True:
        # Compute the next elements of the sequence
        a, b = b, (a + b) % 10

        # Check if we have returned to the original pair
        if (a, b) == starting_pair:
            return True

    return False

# Iterate through each possible pair of numbers from 0 to 9
for i in range(10):
    for j in range(10):
        if seq3(i, j):
            print(f"The starting values ({i},{j}) form a bracelet")
```

## Length of bracelet

This code calculates the length of the bracelet starting with (1,5).

```python
def bracelet_length(a, b):
    # Store the original pair
    starting_pair = (a, b)

    # Initialise the length of the bracelet
    length = 0

    while True:
        # Compute the next elements of the sequence
        a, b = b, (a + b) % 10

        # Increase the length of the bracelet
        length += 1

        # Check if we have returned to the starting pair
        if (a, b) == starting_pair:
            return length

# Calculate the length 
print('The length of the bracelet starting with (1,5) is:',bracelet_length(1, 5))
```

This code calculates all possible bracelet lengths.

```python
# Initialise a set to store the lengths of the bracelets
bracelet_lengths = set()

# Iterate through each possible pair of numbers from 0 to 9
for i in range(10):
    for j in range(10):
        # Calculate the length of the bracelet
        length = bracelet_length(i, j)

        # If the length has not already been calculated, add it to the set
        bracelet_lengths.add(length)

# Print the possible bracelet lengths
print("Possible bracelet lengths are:", sorted(list(bracelet_lengths)))
```
